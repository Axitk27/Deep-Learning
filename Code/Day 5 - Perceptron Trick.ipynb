```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_classification
from matplotlib import pyplot as plt
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data\placement.csv")
```

```python
X,y = make_classification(n_samples=100, n_features=2, n_classes=2, n_redundant=0, n_informative=1, random_state=41, n_clusters_per_class=1, hypercube=False,class_sep=10)
```

```python
plt.Figure(figsize=(10,6))
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
```

```python
def step_function(x):
    return 1 if x >= 0 else 0   
```

```python
def perceptron(X,y):
    X = np.insert(X, 0, 1, axis=1)      #adding bias term to the input data
    weights = np.ones(X.shape[1])       #initializing the weights to 1
    lr = 0.01
    
    for i in range(1000):
        j = np.random.randint(0, X.shape[0])   #selecting a random data point
        linear_output = np.dot(X[j], weights)  #calculating the linear output
        y_hat = step_function(linear_output)   #applying the step function
        weights = weights + lr*(y[j] - y_hat) * X[j]    #updating the weights
    return weights[0], weights[1:]
```

```python
intercept_, coefficients_ = perceptron(X,y)
```

```python
print("Intercept:", intercept_)
print("Coefficients:", coefficients_)
```

```python
m = -coefficients_[0]/coefficients_[1]
b = -intercept_/coefficients_[1]
```

```python
plt.Figure(figsize=(10,6))
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
x = np.linspace(-1, 1, 100)
z = m*x + b
plt.plot(x, z, color='red')
plt.show()
```
