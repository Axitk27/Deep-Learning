```python
from sklearn.datasets import make_classification
from matplotlib import pyplot as plt    
import numpy as np
import seaborn as sns   
```

```python
X,y = make_classification(n_samples=100, n_features=2, n_classes=2, n_redundant=0, n_informative=1, random_state=41, n_clusters_per_class=1, hypercube=False,class_sep=10)
```

```python
plt.Figure(figsize=(10,6))
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
plt.show()
```

```python
def perceptron_loss(X,y):
    w1=w2=b=1
    lr = 0.1
    
    for j in range(10000):
        for i in range(X.shape[0]):
            
            z = w1*X[i][0] + w2*X[i][1] + b
            
            if y[i]*z < 0:
                w1 = w1 + lr*y[i]*X[i][0] 
                w2 = w2 + lr*y[i]*X[i][1] 
                b = b + lr*y[i]
    
    return w1,w2,b
        
```

```python
w1,w2,b
```

```python
w1,w2,b = perceptron_loss(X,y)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)

# x-values for the line
p = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)

# corresponding y-values on the decision boundary
q = -(w1 * p + b) / w2

plt.plot(p, q, color='red', label='Decision Boundary')
plt.legend()
plt.show()
```
