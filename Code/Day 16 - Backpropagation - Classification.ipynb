```python
def initilize_parameter(layer_dims):
    np.random.seed(3)
    parameters={}
    L = len(layer_dims)
    
    for l in range(1,L):
        parameters["W" + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
        parameters["b" + str(l)] = np.zeros((layer_dims[l], 1))

    return parameters
        
```

```python
initilize_parameter([2,2,1])
```

```python
def sigmoid(Z):
    A = 1/(1+np.exp(-Z))
    
    return A
```

```python
def linear_forward(A_prev,W,b):
    Z = np.dot(W.T,A_prev) + b
    return sigmoid(Z)
```

```python
def L_layer_forward(X,parameters):
    
    A = X
    L = len(parameters)//2
    
    for l in range(1,L+1):
        A_prev = A
        Wl = parameters["W" + str(l)]
        bl = parameters["b" + str(l)]
        
        # print("A" + str(l-1) + ": ", A_prev )
        # print("W" + str(l) + ": ", Wl )
        # print("b" + str(l) + ": ", bl)
        # print("--"*20)
        A = linear_forward(A_prev,Wl,bl)
        # print("A" + str(l) + ": ", A)
        # print("**"*20)
        
    return A, A_prev
```

```python
parameters = initilize_parameter([2,2,1])
```

```python
X = df[["cgpa","profile_score"]].values[2].reshape(2,1)
y = df[["is_placed"]].values[2][0]
```

```python
y_hat, A1 = L_layer_forward(X,parameters)
```

```python
y_hat = y_hat[0][0]
```

```python
A1
```

```python
def loss_calculation(y,y_hat):
    return (-y * np.log(y_hat) - (1-y) * np.log(1 - y_hat))
```

```python
loss_calculation(y,y_hat)
```

```python
def update_parameters(parameters,y,y_hat,A1,X, n):
    parameters["W1"][0][0] =  parameters["W1"][0][0] + (n * (y-y_hat) * parameters["W2"][0][0] * A1[0][0] * (1-A1[0][0]) * X[0][0])
    parameters["W1"][0][1] =  parameters["W1"][0][1] + (n * (y-y_hat) * parameters["W2"][0][0] * A1[0][0] * (1-A1[0][0]) * X[0][0])    
    parameters["b1"][0][0] =  parameters["b1"][0][0] + (n * (y-y_hat) * parameters["W2"][0][0] * A1[0][0] * (1-A1[0][0]))    
    
    parameters["W1"][1][0] =  parameters["W1"][1][0] + (n * (y-y_hat) * parameters["W2"][1][0] * A1[1][0] * (1-A1[1][0]) * X[0][0])
    parameters["W1"][1][1] =  parameters["W1"][1][1] + (n * (y-y_hat) * parameters["W2"][1][0] * A1[1][0] * (1-A1[1][0]) * X[1][0])    
    parameters["b1"][0][0] =  parameters["b1"][0][0] + (n * (y-y_hat) * parameters["W2"][1][0] * A1[1][0] * (1-A1[1][0]))  

    parameters["W2"][0][0] =  parameters["W2"][0][0] + (n * (y-y_hat) * A1[0][0])
    parameters["W2"][1][0] =  parameters["W2"][1][0] + (n * (y-y_hat) * A1[1][0])    
    parameters["b2"][0][0] =  parameters["W2"][1][0] + (n * (y-y_hat))    
```

```python
update_parameters(parameters,y,y_hat,A1,X,0.01)
```

```python
# epoch implementation 


parameters = initilize_parameter([2,2,1])
epochs = 50


for i in range(epochs):
    Loss = []
    
    for j in range(df.shape[0]):
        X = df[["cgpa","profile_score"]].values[j].reshape(2,1)
        y = df[["is_placed"]].values[j][0]
        y_hat, A1 = L_layer_forward(X,parameters)
        y_hat = y_hat[0][0]
        Loss.append(loss_calculation(y,y_hat))
        update_parameters(parameters,y,y_hat,A1,X,0.01)
    
    print("Epoch - ", i, "Loss - ", np.array(Loss).mean())  
```
